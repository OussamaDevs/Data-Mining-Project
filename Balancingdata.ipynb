```python
import pandas as pd
df = pd.read_csv("diabetes_engineered_normalized.csv")

df.head()
```

```python
from sklearn.utils import resample
# Calculate the total number of samples
total_samples = len(df)

# Calculate the target number of diabetic cases to achieve 13%
target_diabetic_count = int(total_samples * 0.13)

# Separate the diabetic and non-diabetic classes
diabetic_samples = df[df['diabetes'] == 1]
non_diabetic_samples = df[df['diabetes'] == 0]

# If diabetic samples are less than the target, we need to increase them
# If diabetic samples are more than the target, we need to reduce them

if len(diabetic_samples) < target_diabetic_count:
    # Increase the diabetic samples to match the target using resampling
    diabetic_samples = resample(diabetic_samples,
                                replace=True,  # Sampling with replacement
                                n_samples=target_diabetic_count,  # Exactly the target count
                                random_state=42)
else:
    # Reduce the diabetic samples to match the target
    diabetic_samples = diabetic_samples.sample(n=target_diabetic_count, random_state=42)

# Now calculate the number of non-diabetic samples
non_diabetic_count = total_samples - target_diabetic_count

# Sample the non-diabetic class to match the remaining space
non_diabetic_samples = non_diabetic_samples.sample(n=non_diabetic_count, random_state=42)

# Combine the balanced diabetic and non-diabetic classes
df_balanced = pd.concat([diabetic_samples, non_diabetic_samples])

# Step 5: Calculate and print the new class distribution in percentage
class_percentage = df_balanced['diabetes'].value_counts(normalize=True) * 100
print(f"New class distribution (in percentage):\n{class_percentage}")
```

```python
df_balanced.to_csv('diabetes_engineered_normalized_balanced.csv', index=False)
```
