```python
import pandas as pd

df = pd.read_csv("diabetes_engineered_normalized_balanced.csv")
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder

# Assuming df is already defined

# Initialize label encoder
label_encoder = LabelEncoder()

# Convert 'smoking_history' into numerical values
df['smoking_history'] = label_encoder.fit_transform(df['smoking_history'])

# Convert 'age_group' into numerical values
df['age_group'] = label_encoder.fit_transform(df['age_group'])

# Convert 'gender' columns (binary: Female, Male, Other) into numerical values
df['gender_Female'] = df['gender_Female'].astype(int)
df['gender_Male'] = df['gender_Male'].astype(int)
df['gender_Other'] = df['gender_Other'].astype(int)

# Convert 'bmi_category' into numerical values
df['bmi_category'] = label_encoder.fit_transform(df['bmi_category'])

# Display the first few rows of the transformed dataframe to verify
print(df.head())
```

```python
df.to_csv('diabetes_engineered_normalized_balanced_allNumerical.csv', index=False)
```
